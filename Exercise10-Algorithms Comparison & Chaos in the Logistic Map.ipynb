## Question 1 (Algorithm instability of Euler Method)

```python
#we want to show how Algorithm instability will increase if we enlargen the steps (from exercise 9)
#the Discharging equation of a capacitor with Euler method:

import matplotlib.pyplot as plt
import numpy as np

step=0.01 #step size
st=1 #simulation time

def f(V): #dv/dt=f(v) -> dv/dt=-v/RC -> f(v)=-V
    return -V

v=10 #initial voltage
V=[] 
t=[]
for i in range(0,int(st/step),2):
    vn1=v
    v+=step*f(v)  # Euler
    vn=v
    v+=step*f(v)
    V.append(v-step*(f(vn1)-f(vn))) 
    t.append(step*i)
plt.plot(t,V,label='Euler method',linewidth=3) 


#to see how close this method is to the analytic method we got to solve it analytically
# dv/dt=-v we guess the solution as v(t)=Be^(At) to find A: we put it to the equetion -> Ae^(At)=-e^(At) -> A=-1 so the solution is v=Be^(-t)
# due to initial state: B=10 
x=np.linspace(0,st,100)
y=10*np.exp(-x)
plt.plot(x,y,label='Analytic method')
plt.legend()
plt.xlabel("t")
plt.ylabel("V")
plt.show()
```

### 1. Euler method

#### step=0.01

```python
w =1.0#angular frequency
step=0.01
T=10
x=1
v= 0

X=[]
V=[]
t=[]
E=[]
#Euler
for i in range(int(T/step)): 
    x+=step*v
    v+=-step*(w**2)*x
    X.append(x)
    V.append(v)
    t.append(i*step)
    E.append(0.5*v**2+0.5*(w**2)*x**2) #E=1/2 v^2 +1/2 w^2 x^2

plt.plot(t,X)
plt.title(f"Euler(x-t), step={step}")
plt.xlabel("t")
plt.ylabel("X")
plt.show()

plt.plot(X,V)
plt.title(f"Euler Phase Diagram, step={step}")
plt.xlabel("X")
plt.ylabel("V")
plt.axis('equal')
plt.show()

plt.plot(t, E)
plt.title(f"Total Energy(Euler), step={step}")
plt.xlabel("t")
plt.ylabel("E")
plt.show()

print()

# Analytic 
xa=np.linspace(0,10,1000)
ya=np.cos(xa)
plt.plot(xa,ya)
plt.title(f"Analytic (x-t), step={step}")
plt.xlabel("t")
plt.ylabel("X")
plt.show()

ta=np.arange(0, T, step)
xap=np.cos(w * ta)
vap=-w*np.sin(w * ta)
plt.plot(xap, vap)
plt.title(f"Analytic Phase Diagram, step={step}")
plt.xlabel("X")
plt.ylabel("V")
plt.axis('equal')
plt.show()
```

### 3. leapfrog method

#### step=0.01

```python
import numpy as np
import matplotlib.pyplot as plt

w=1 #angular frequency
step=0.01
T=10
X=[]
V=[]
E=[]
t=[]     
x =1                        
vhalf=-0.5*step*x*w**2 #v half= 1/2 h x w^2

#leapfrog integration
for n in range(int(T/step)):            
    vhalf+=-step*w**2*x
    x+=step*vhalf                            
    v=vhalf-0.5*step*x*w**2 #vhalf=1/2 h xw^2
    X.append(x)
    V.append(v)
    E.append(0.5*(v**2+w**2*x**2))
    t.append(n*step)

    
plt.plot(t,X)
plt.title(f"Leapfrog(x-t),step={step}")
plt.xlabel("t")
plt.ylabel("X")
plt.show()

plt.plot(X,V)
plt.title(f"Leapfrog Phase Diagram, step={step}")
plt.xlabel("X")
plt.ylabel("V")
plt.axis('equal')
plt.show()

plt.plot(t, E)
plt.title(f"Total Energy(Leapfrog)step={step}")
plt.xlabel("t")
plt.ylabel("E")
plt.show()


print()

# Analytic 
xa=np.linspace(0,10,1000)
ya=np.cos(xa)
plt.plot(xa,ya)
plt.title(f"Analytic (x-t), step={step}")
plt.xlabel("t")
plt.ylabel("X")
plt.show()

ta=np.arange(0, T, step)
xap=np.cos(w * ta)
vap=-w*np.sin(w * ta)
plt.plot(xap, vap)
plt.title(f"Analytic Phase Diagram, step={step}")
plt.xlabel("X")
plt.ylabel("V")
plt.axis('equal')
plt.show()
```

### 4. Velocity Verlet

#### step=0.01

```python
t=[0, 1]
X=[]
V=[]
E=[]

w=1
step=0.01
T=10
x=1
v=0
a=-x*w**2  #initial acceleration from Hooke's law: a = -x
A=[a]
X.append(x)

#using taylor expansion to make x
x=X[0]+step*v-0.5*step**2*X[0]*w**2
X.append(x)
V.append(v)

#Verlet
for n in range(2, int(T/step)):
    x=2*X[-1]-X[-2]-X[-1]*step**2*w**2 # this method needs its two last steps thats why i made the two x before
    X.append(x)
    t.append(n*step)

for n in range(1,int(T/step)-1):
    V.append((X[n+1]-X[n-1])/(2*step))

for i in range(len(V)):
    E.append(0.5*V[i]**2+0.5*w**2*X[i]**2)  #E=1/2 v^2+1/2w^2 x^2



plt.plot(t,X)
plt.title(f"Velocity Verlet(x-t), step={step}")
plt.xlabel("t")
plt.ylabel("X")
plt.show()

plt.plot(X[:-1],V)
plt.title(f"Velocity Verlet Phase Diagram, step={step}")
plt.xlabel("X")
plt.ylabel("V")
plt.axis('equal')
plt.show()

plt.plot(t[:-1], E)
plt.title(f"Total Energy(Velocity Verlet), step={step}")
plt.xlabel("t")
plt.ylabel("E")
plt.show()

print()

# Analytic 
xa=np.linspace(0,10,1000)
ya=np.cos(xa)
plt.plot(xa,ya)
plt.title(f"Analytic (x-t), step={step}")
plt.xlabel("t")
plt.ylabel("X")
plt.show()

ta=np.arange(0, T, step)
xap=np.cos(w * ta)
vap=-w*np.sin(w * ta)
plt.plot(xap, vap)
plt.title(f"Analytic Phase Diagram, step={step}")
plt.xlabel("X")
plt.ylabel("V")
plt.axis('equal')
plt.show()
```

### 5. Beeman's method

#### step=0.01

```python
t=[]
X=[]
V=[]
A=[]
E=[]

w=1
step=0.01
T=10
x=1
v=0
a=-x*w**2  #initial acceleration from Hooke's law: F =-kx=-wx
A=[a]
X.append(x)
V.append(v)
A.append(a)

# Estimate x using Taylor expansion
x = x + step * v + 0.5 * step**2 * a
X.append(x)
a=-w**2*x
A.append(a)

# Estimate v using Beeman's velocity
v=v+0.5*step*(A[-1] + A[-2])
V.append(v)

# Beeman integration loop starts from n=1
for i in range(1, int(T/step)):
    x=x+step*v+(step**2/6)*(4*A[-1]-A[-2])
    a=-w**2*x #acceleration in predicted x
    A.append(a)
    
    v=v+(step/6)*(2*A[-1]+5*A[-2]-A[i-3])
    
    X.append(x)
    t.append(i*step)
    V.append(v)
    E.append(0.5*v**2+0.5*(w**2)*x**2) #E=1/2 v^2+1/2w^2 x^2


plt.plot(t,X[:-2])
plt.title(f"Beeman(x-t), step={step}")
plt.xlabel("t")
plt.ylabel("X")
plt.show()

plt.plot(X,V)
plt.title(f"Beeman Phase Diagram, step={step}")
plt.xlabel("X")
plt.ylabel("V")
plt.axis('equal')
plt.show()

plt.plot(t, E)
plt.title(f"Total Energy(Beeman), step={step}")
plt.xlabel("t")
plt.ylabel("E")
plt.show()

print()

# Analytic 
xa=np.linspace(0,10,1000)
ya=np.cos(xa)
plt.plot(xa,ya)
plt.title(f"Analytic (x-t), step={step}")
plt.xlabel("t")
plt.ylabel("X")
plt.show()

ta=np.arange(0, T, step)
xap=np.cos(w*ta)
vap=-w*np.sin(w*ta)
plt.plot(xap, vap)
plt.title(f"Analytic Phase Diagram, step={step}")
plt.xlabel("X")
plt.ylabel("V")
plt.axis('equal')
plt.show()
```

## Question 3 (Chaos)

```python
import numpy as np
import matplotlib.pyplot as plt

# Logistic map function
x0=0.5
iterations=1000
plotting = 300  #plotting the only "plotting" numbers because the other first will be so close to the initial point and chaos won't be seen
R=np.linspace(x0, 1, 10000) 

def logisticmap(x, r):
    return 4*r*x*(1- x)


plt.figure(figsize=(10, 6)) #to make the graph bigger

for r in R:
    x=x0 #restarting x so we will do the chaos from the initial point
    #getting far away from the initial point
    for i in range(iterations - plotting):  # the one that are not plotted
        x =logisticmap(x, r)
    X=[]
    #plotting the one that are far from the initial point
    for j in range(plotting):
        x= logisticmap(x, r)
        X.append(x)
    plt.plot([r]*plotting, X, ',k' ,alpha=0.1) #alpha is for the width of the graph

plt.title("Bifurcation Diagram of Logistic Map")
plt.xlabel("Control parameter r")
plt.ylabel("x")
plt.show()

#approximating bifurcation points (stimated from the plot)
r1= 0.75
r2 =0.88
r3=0.89
r4 =0.9

Feigenbaum1=(r2-r1) / (r3-r2)
Feigenbaum2=(r3-r2) / (r4-r3)

print(f"first Feigenbaum constant: {Feigenbaum1:.2f}")
print(f"second Feigenbaum constant: {Feigenbaum2:.2f}")

```
